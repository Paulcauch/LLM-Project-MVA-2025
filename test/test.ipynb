```python
import torch
import torch.nn as nn
import torch.optim as optim

import numpy as np

import matplotlib.pyplot as plt

from src.rope import RoPEPositionalEncoding
```

```python
rope = RoPEPositionalEncoding(1)
out = rope(torch.tensor([i for i in range(20)]))

x_values = out[:, 0]
y_values = out[:, 1]
indices = np.arange(len(x_values))
for i, (x, y) in enumerate(zip(x_values, y_values)):
    plt.text(x, y, str(i+1), fontsize=10)

plt.scatter(x_values, y_values, c=indices, cmap='viridis')
plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.show()
```
